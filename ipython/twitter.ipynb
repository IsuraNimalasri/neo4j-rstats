## Cypher Magic

To use Cypher magic, you'll need the [ipython-cypher](https://github.com/versae/ipython-cypher) package:

`pip install ipython-cypher`

```python
%load_ext cypher
%matplotlib inline
```

```python
%%cypher
MATCH (u:User)-[:POSTS]->(t:Tweet)
RETURN u.username, t.text
LIMIT 5
```

```python
%%cypher
MATCH (t:Tweet)-[:MENTIONS]->(u:User)
RETURN u.username, COUNT(t) AS mentions
ORDER BY mentions DESC
LIMIT 5
```

```python
%%cypher
MATCH (h:Hashtag)-[:TAGS]->(t:Tweet)
RETURN h.name AS hashtag, COUNT(t) AS tweets
ORDER BY tweets DESC
LIMIT 5
```

## Plotting

You'll need the latest (development) version of ggplot if you want to have any luck with it:

`pip install git+git://github.com/yhat/ggplot.git`

```python
from ggplot import *

results = %cypher MATCH (h1:Hashtag)-[:TAGS]->(t:Tweet) \
                  WITH h1, COUNT(t) AS count \
                  ORDER BY count DESC \
                  LIMIT 10 \
                  \
                  WITH COLLECT(h1) AS hashtags \
                  UNWIND hashtags AS h1 \
                  UNWIND hashtags AS h2 \
                  \
                  MATCH (h1)-[:TAGS]->(:Tweet)<-[:TAGS]-(h2) \
                  RETURN h1.name, h2.name, COUNT(*) AS count
                    
df = results.get_dataframe()
df[:5]
```

```python
ggplot(df, aes('h1.name', 'h2.name')) + geom_tile(aes(fill='count'))
```

Fun with some guy's wordcloud package:

`pip install git+git://github.com/amueller/word_cloud.git`

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import StringIO

result = %cypher MATCH (t:Tweet) \
                 WHERE HAS(t.text) \
                 RETURN t.text AS text

text = ""

for row in result:
    text += " " + row[0].encode('utf-8')
    
text = text.replace("RT ", "")
text = text.replace("https://", "")
text = text.replace(".co", "")

f = StringIO.StringIO(text)
text = f.read()
words = WordCloud(width=1920, height=1080, font_path='/Library/Fonts/verdana.ttf')
wordcloud = words.generate(text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## NetworkX

`pip install networkx`

```python
import networkx as nx

results = %cypher MATCH p = (:Hashtag)-[:TAGS]->(:Tweet)<-[:TAGS]-(:Hashtag) \
                  RETURN p \
                  LIMIT 10

g = results.get_graph()
plt.figure(figsize=(6,4));
nx.draw(g)
```

```python
node_map = {'Hashtag':'#FF0099', 'Tweet':'#660066'} 

g = results.get_graph()
plt.figure(figsize=(6,4));
nx.draw(g, node_color=[node_map[g.node[node]['labels'][0]] for node in g])
```

```python
nx.draw(g, node_color=[node_map[g.node[node]['labels'][0]] for node in g],
           node_size=[x * 100 for x in g.degree().values()])
```
